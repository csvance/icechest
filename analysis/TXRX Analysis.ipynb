```python
data = open("RX_ACTUATED.txt", "r").read()

print(len(data.split("\n")))

for row in data.split("\n")[1:]:
    
    if len(row) == 0:
        continue 
        
    cols = row.split(",")
    if cols[1] == '\\r (0x0D)':
        stack = []
    elif cols[1] == '\\n (0x0A)':
        print(" ".join(stack))
    else:
        stack.append(cols[1].split(' ')[0])

        
```
